```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv')

print(df)
```

```python
x = cdf.ENGINESIZE.to_numpy()
y = cdf.CO2EMISSIONS.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.2, random_state=42)
train_x
```

```python
train_y
```

```python
type(train_x), np.shape(train_x), np.shape(train_y)
```

```python
type(test_x), np.shape(test_x), np.shape(test_y)
```

```python
from sklearn import linear_model

# create a model object
regressor = linear_model.LinearRegression()

regressor.fit(train_x.reshape(-1, 1), train_y)
```

```python
# Print the coefficients
print ('Coefficients: ', regressor.coef_[0]) # with simple linear regression there is only one coefficient, here we extract it from the 1 by 1 array.
print ('Intercept: ',regressor.intercept_)
```

```python
plt.scatter(train_x, train_y)
plt.plot(train_x, regressor.coef_ * train_x + regressor.intercept_, '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Use the predict method to make test predictions
pred_y = regressor.predict(test_x.reshape(-1,1))

# Evaluation
print("Mean absolute error: %.2f" % mean_absolute_error(test_y, pred_y))
print("Mean squared error: %.2f" % mean_squared_error(test_y, pred_y))
print("Root mean squared error: %.2f" % np.sqrt(mean_squared_error(test_y, pred_y)))
print("R2-score: %.2f" % r2_score(test_y, pred_y))
```

```python
plt.scatter(test_x, test_y)
plt.plot(test_x, regressor.coef_ * test_x + regressor.intercept_, '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```
